```python
import pandas as pd
import os


file_path = r"C:\Users\Balaganesh S\Desktop\employee_data_updated.xlsx"


original_file_path = r"C:\Users\Balaganesh S\Downloads\employee data.xlsx"

if not os.path.exists(original_file_path):
    print("Original file does not exist.")
else:
    try:
        
        dataset = pd.read_excel(original_file_path)

        print("Existing Data:")
        print(dataset.head())

        
        new_data1 = {
            'Employee': ['Alice', 'Bob', 'Agila', 'Bala', 'Ramya', 'Surya', 'Revin'],  
            'Location': ["London", "New York", "Tokyo", "India", "United States", "Brazil", "Nigeria"],
            'Sex': ['Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female'],  
            'Salary': [63850, 55650, 69870, 80000, 85000, 90000, 65000],  
        }

        
        new_data2 = {
            'Employee': ['Ganesh', 'Vinu', 'Mani'],
            'Location': ['US', 'Paris', 'Canada'],
            'Sex': ['Male', 'Female', 'Male'],
            'Salary': [80000, 75000, 70000]
        }

        
        new_df1 = pd.DataFrame(new_data1)
        new_df2 = pd.DataFrame(new_data2)

        
        with pd.ExcelWriter(file_path, engine='openpyxl', mode='w') as writer:
            
            dataset.to_excel(writer, sheet_name='Sheet1', index=False)
            
            
            new_df1.to_excel(writer, sheet_name='Sheet2', index=False)

            
            new_df2.to_excel(writer, sheet_name='Sheet3', index=False)

        print("Data added successfully.")

        
        sheet2_data = pd.read_excel(file_path, sheet_name='Sheet2')
        sheet3_data = pd.read_excel(file_path, sheet_name='Sheet3')
        
        print("\nData in Sheet2:")
        print(sheet2_data)
        
        print("\nData in Sheet3:")
        print(sheet3_data)

    except Exception as e:
        print(f"An error occurred: {e}")
```
